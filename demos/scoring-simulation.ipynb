# Lending Risk Scoring Simulation
This notebook simulates threshold logic for a hypothetical ML-based risk scoring system.

```python
import numpy as np
import pandas as pd

# simulate probability of default scores
np.random.seed(42)
scores = np.random.beta(2, 10, 500)
df = pd.DataFrame({'pd_score': scores})

# thresholds
LOW = 0.02
HIGH = 0.08

def apply_decision(score):
    if score < LOW:
        return 'approve'
    elif score > HIGH:
        return 'decline'
    else:
        return 'manual_review'

df['decision'] = df['pd_score'].apply(apply_decision)
df.head()
```
